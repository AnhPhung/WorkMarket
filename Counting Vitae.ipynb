```python
import re
import matplotlib.pyplot as plt
%matplotlib inline
import string
import numpy as np
```

```python
with open('Resume Anh Phung.txt', 'r') as f:
    txt=f.read()
```

```python
punc = string.punctuation
pattern1 = re.compile('[{punc}]'.format(punc = re.escape(punc)))
pattern2 = re.compile(r'[\s]|\\\n|\\\t|â€“')
txt = re.sub(pattern1, '', txt)
txt = re.sub(pattern2, '', txt)
txt=txt.lower()
```

```python
char = []
count = {}
for i in txt:
    if i not in char:
        count[i] = 1
        char.append(i)
    elif i in char:
        count[i] = count[i]+1
```

```python
pos = np.arange(len(count.keys()))
plt.bar(pos, count.values(), align='center')
plt.xticks(pos, count.keys())
plt.ylabel('Character Frequency')
plt.title('My Resume Character Frequency')
plt.show()
```
